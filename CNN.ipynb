# CNN

The CNN succeeded because not every neuron in a layer is connected to all neurons in a different layer. Besides the hnadling was changed into a scale invariant method.The cnn network trains the values for the kernels. The input layer becomes the amount of pixels.

A pooling layer takes the outcomes of the convulotion layer and applies a kernel over this. In most cases this is a max-pool which selects the highest value.

The stride is how many pixels it moves to the left. Stride 2 is therefore jumping over 1 pixel.

The low level features are learned in the ealrier layers. The higher level features are learned in later layers.

A fully connected layer is applied to all of the result to get the final prediction.

## Applying techniques

### 1d convulotion
1d convulotion is applied to the last convulotion layer such that the depth is reduced to 1 again.

![Screenshot of 1d convulotion](Images/Screenshot%202025-12-02%20at%2015.07.51.png)

### How to reduce overfitting? 

1. Simplify the model (reduce the amount of neurons and layers)
2. Penalize high values for kernels by regularization
3. Dropout, randomly remove some of the neurons in the network during each iteration
