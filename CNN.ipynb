## Applying techniques

### 1d convulotion
1d convulotion is applied to the last convulotion layer such that the depth is reduced to 1 again.

![Screenshot of 1d convulotion](Images/Screenshot%202025-12-02%20at%2015.07.51.png)

### How to reduce overfitting? 

1. Simplify the model (reduce the amount of neurons and layers)
2. Penalize high values for kernels by regularization
3. Dropout, randomly remove some (typically 50%) of the neurons in the network during each training iteration.
4. Batch normalization, normalize (gaussian) the outcome of the activation function.

## Practical aspects 
:::{.callout-tip title="transfer learning"}
In most cases it is actually better to use a pre-trained model eventhough this model is trained on a different item (must be images) because it already knows how to extract features. This is also called transfer learning.
:::

Batch gradient descent can lead to very slow training because it needs to save all the updates of each weight for each image in the dataset. Stochastic gradient descent can be quick however can lead to an osicilating learning curve. Mini-batch gradient descent is a sort of combination of both. These mini-batches should contain each class.
